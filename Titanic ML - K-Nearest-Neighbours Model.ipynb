### Test dataset import and preprocessing

```python
raw_test_data = pd.read_csv('test.csv')
```

```python
raw_test_data
```

```python
test_data = raw_test_data.drop(['PassengerId', 'Name', 'Age', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
```

```python
test_data['Family_size'] = test_data['SibSp'] + test_data['Parch']
```

```python
test_data = test_data.fillna(test_data.mean())
```

```python
X_test = test_data.drop(['SibSp', 'Parch'], axis = 1)
```

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split

X_train_mod, X_test_mod, Y_train_mod, Y_test_mod = train_test_split(X_train, Y_train, test_size=0.20)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train_mod)

X_train_mod = scaler.transform(X_train_mod)
X_test_mod = scaler.transform(X_test_mod)
```

### KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train_mod, Y_train_mod)
```

```python
Y_pred = model.predict(X_test_mod)
```

### Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(Y_test_mod, Y_pred))
print(classification_report(Y_test_mod, Y_pred))
```

```python
Predicted = model.predict(X_test)
```

### Output in csv

```python
output = pd.DataFrame({'PassengerId': raw_test_data.PassengerId, 'Survived': Predicted})
output.to_csv('my_submissionKNN.csv', index=False)
print("Your submission was successfully saved!")
```
